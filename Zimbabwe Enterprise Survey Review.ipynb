## Obstacles to Businesses in Zimbabwe
Understanding constraints to firms is important as the nation prepares to vote for a new government. 

Access to Finance is the most frequently listed as the main obstacle to doing business in Zimbabwe. 22.5% of the firms surveyed list it as a constraint. Close second and third are Practices of Competitors in the Informal Sector (21.5%) and Political Instability (17.5%), respectively. In fact, these top 3 obstacles are listed by 61.5% of all firms surveyed. 

This is consistent with my (limited) knowledge of Zimbabwe. The ongoing cash crisis originally brought about by the debasing of the currency from hyperinflation has led to difficulties in accessing hard currency to facilitate imports.
